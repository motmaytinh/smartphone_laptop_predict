```python
import pandas as pd
import re
```

```python
data = pd.read_csv("laptop_data.csv", sep =",")
data = data.drop('Name', axis = 1)

for index, row in data.iterrows():
    # Price
    data.loc[index, 'Price'] = int(''.join(str(row['Price'])[0:-2].split('.')))#/1000000.0
    # Clock 
    data.loc[index, 'Clock'] = float(str(row['Clock'])[0:-4])
    # Cache
    data.loc[index, 'Cache'] = int(str(row['Cache'])[0:-2])
    # Ram
    data.loc[index, 'Ram'] = int(str(row['Ram'])[0:-2])
    # OS
    if data.loc[index, 'OS'] == 'free dos' or data.loc[index, 'OS'] == 'không có':
        data.loc[index, 'OS'] = 0
    elif data.loc[index, 'OS'] == 'os x':
        data.loc[index, 'OS'] = 1
    elif data.loc[index, 'OS'] == 'linux' or data.loc[index, 'OS'] == 'ubuntu':
        data.loc[index, 'OS'] = 2
    else:
        data.loc[index, 'OS'] = 3
#     # Max ram
#     data.loc[index, 'Max Ram'] = data.loc[index, 'Ram'] if (data.loc[index, 'Max Ram'] == "Đang cập nhật" or data.loc[index, 'Max Ram'] == "Không") else int(str(data.loc[index, 'Max Ram'])[0:-2])
    # Display Size
    data.loc[index, 'Display Size'] = float(str(row['Display Size'])[0:-1])
    # Display Res
    disres = re.search('(\d)+ x (\d)+', str(row['Display Res']))
    if disres != None:
        res = disres.group(0).split(' x ')
        if res != []:
            data.loc[index,'displayWidth'] =  min(int(res[0]),int(res[1]))
            data.loc[index,'displayHeight'] =  max(int(res[0]),int(res[1]))
        else: 
            data.loc[index,'displayWidth'] =  -1
            data.loc[index,'displayHeight'] =  -1
    else:
        data.loc[index,'displayWidth'] =  -1
        data.loc[index,'displayHeight'] =  -1
    # VGA Memory
    data.loc[index, 'VGA Memory'] = 0 if data.loc[index, 'VGA Memory'] == 'Không' else int(str(data.loc[index, 'VGA Memory'])[0:-2])
    # HDD
    if data.loc[index, 'HDD'] == 'Không có':
        data.loc[index, 'HDD'] = 0
    elif int(str(data.loc[index, 'HDD'])[0:-2]) == 1:
        data.loc[index, 'HDD'] = 1000
    elif int(str(data.loc[index, 'HDD'])[0:-2]) == 2:
        data.loc[index, 'HDD'] = 2000
    else:
        data.loc[index, 'HDD'] = int(str(data.loc[index, 'HDD'])[0:-2])
    # SSD
    data.loc[index, 'SSD'] = 0 if data.loc[index, 'SSD'] == 'Không có' else int(str(data.loc[index, 'SSD'])[0:-2])
    # Cells pin
    if row['Cells pin'] == 'Gắn liền':
        data.loc[index, 'Cells pin'] = 6
    # Weight
    wn1 = re.findall('\d+\.+\d+', data.loc[index, 'Weight'])
    
    if wn1:
        data.loc[index, 'Weight'] = float(wn1[0])
    else:
        wn2 = re.findall('\d+', data.loc[index, 'Weight'])
        data.loc[index, 'Weight'] = float(wn2[0])

# Brand
data['Brand'] = data['Brand'].str.lower()
data['Brand'] = data['Brand'].astype('category')
# Purpose
data['Purpose'] = data['Purpose'].str.lower()
data['Purpose'] = data['Purpose'].astype('category')
# Color
data['Color'] = data['Color'].str.lower()
data['Color'] = data['Color'].astype('category')
# CPU Series + Gen
data['gen'] = data['gen'].str.lower()
data['gen'] = data['gen'].astype('category')
# CPU
data['CPU'] = data['CPU'].str.lower()
data['CPU'] = data['CPU'].astype('category')
# OS
# data['OS'] = data['OS'].str.lower()
# data['OS'] = data['OS'].astype('category')
# Touchable
data['Touchable'] = data['Touchable'].str.lower()
data['Touchable'] = data['Touchable'].astype('category')
# Integrated Graphics
data['Integrated Graphics'] = data['Integrated Graphics'].str.lower()
data['Integrated Graphics'] = data['Integrated Graphics'].astype('category')
# Add Graphics
data['AddGraphics'] = data['AddGraphics'].str.lower()
data['AddGraphics'] = data['AddGraphics'].astype('category')
# Optical Disk
data['Optical Disk'] = data['Optical Disk'].str.lower()
data['Optical Disk'] = data['Optical Disk'].astype('category')
# Number Keys
data['Number Keys'] = data['Number Keys'].str.lower()
data['Number Keys'] = data['Number Keys'].astype('category')
# LED keys
data['LED'] = data['LED'].str.lower()
data['LED'] = data['LED'].astype('category')
# LAN Ports
data['LANPorts'] = data['LANPorts'].str.lower()
data['LANPorts'] = data['LANPorts'].astype('category')
# VGA Ports
data['VGAPorts'] = data['VGAPorts'].str.lower()
data['VGAPorts'] = data['VGAPorts'].astype('category')


data = data.drop('Display Res', axis = 1)
data = data.drop('Max Ram', axis = 1)
data = data.drop('VR support', axis=1)
data = data.drop('Integrated Graphics', axis=1)
data = data.drop('Warranty', axis=1)
```
